```python
# create function to retrive weight and value list
def get_matrix(file_path="Knapsack_Instances.csv"):
    df = pd.read_csv(file_path, header = None)
    df.reset_index(inplace = True, drop = True)
    sizes_list=[list([int(num) for num in df.loc[i, 0:49]]) for i in range(0,1000,2)]
    rewards_list=[list(df.loc[i+1, 0:49]) for i in range(0,1000,2)]
    
    return sizes_list,rewards_list
```

```python
# create function to return sum of rewards
def Eva(sizes,rewards,solution,C):
    current_obj = 0 
    total_size = np.dot(sizes,solution)
    total_reward = np.dot(rewards,solution)
    if total_size <= C:
        current_obj+=total_reward
    return current_obj
```

```python
# create function to return solution list and dictionary
def individual(weight_list, value_list, max_weight):

    solu_list = [0] * (max_weight + 1)

    solu_dict = {i: [] for i in range(0, max_weight + 1)}

    for item in range(len(weight_list)):

        list_copy = solu_list.copy()
        dict_copy = solu_dict.copy()
        for weight in range(max_weight + 1):

            if weight >= weight_list[item]:
      
                list_copy[weight] = max(solu_list[weight - weight_list[item]] + value_list[item], solu_list[weight])
      
                if list_copy[weight] > solu_list[weight]:
                    dict_copy[weight] = [item]
                    for item_before in solu_dict[weight - weight_list[item]]:
                        dict_copy[weight].append(item_before)

        solu_dict = dict_copy.copy()
        solu_list = list_copy.copy()

    return solu_list,solu_dict
```

```python
# create function to return sum of rewards for random C between [100,150]
def Evaluate(sizes_list,rewards_list,solution_list):
    
    knap_sizes = np.random.uniform(100,150,len(sizes_list))
    current_obj = 0 
    
    for i in range(len(rewards_list)):
        sizes=sizes_list[i]
        rewards=rewards_list[i]
        solution = solution_list[i]
        C = knap_sizes[i]
        total_size = np.dot(sizes,solution)
        total_reward = np.dot(rewards,solution)
        if total_size <= C:
            current_obj+=total_reward
    return current_obj
```

```python
# Dynamic programming approach to achieve largest value
weight_list = get_matrix()[0]
value_list = get_matrix()[1]

final_weight=[]
final_solution=[]
for i in range(len(weight_list)):
    #assume bag size is 100
    bag_size=100
    solu=0
    solu_update=[0]*50
    dict=individual(weight_list[i], value_list[i], 150)[1]
    large_weight=0
    update=[]
    for weight in range(100,150):
        solution_list=[1 if i in dict[weight] else 0 for i in list(range(50))]

        new_list=[Eva(weight_list[i],value_list[i],solution_list,n) for n in range(100,150)]
        new_solu=sum(new_list)/len(new_list)
        update+=[new_solu]
        # update solution if new is larger
        if new_solu>solu:
            bag_size=weight
            solu=new_solu
            solu_update=solution_list
            large_weight=weight

    final_weight.append(solu)
    final_solution.append(solu_update)
    print(i,bag_size,solu,solu_update)
print(sum(final_weight))
```

```python
# create function to generate csv output file
def output_csv(sizes_list,rewards_list,solution_list,output_name="145_midterm_knapsack.csv"):
    f = open(output_name, "w", newline='')
    csv_writer = csv.writer(f)
    
    for i in range(len(solution_list)):
        csv_writer.writerow(sizes_list[i])
        csv_writer.writerow(rewards_list[i])
        csv_writer.writerow(solution_list[i])
    f.close()
```

```python
# Evalutation of output solution
submitted_file_path = "145_midterm_knapsack.csv"

def Evaluate_Sol(file_path):
    student_id = file_path.split("_")[0]
    df_solution = pd.read_csv(file_path, header = None)
    df_solution.reset_index(inplace = True, drop = True)
    current_obj = 0 

    for i in range(0,1500,3):
        sizes = list(df_solution.loc[i, 0:49])
        rewards = list(df_solution.loc[i+1, 0:49])
        solution = [bool(i) for i in list(df_solution.loc[i+2, 0:49])]
        C = knap_sizes[int(i/3)]
 
        total_size = np.dot(sizes,solution)
        total_reward = np.dot(rewards,solution)
        if total_size <= C:
            current_obj+=total_reward
    return {student_id: current_obj}


final_results = Evaluate_Sol(submitted_file_path)
final_results
```
