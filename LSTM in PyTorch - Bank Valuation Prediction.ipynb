```python
# scale train and test data to [-1, 1]
def scale(train, test):
	# fit scaler
	scaler = MinMaxScaler(feature_range=(-1, 1))
	scaler = scaler.fit(train)
	# transform train
	train = train.reshape(train.shape[0], train.shape[1])
	train_scaled = scaler.transform(train)
	# transform test
	test = test.reshape(test.shape[0], test.shape[1])
	test_scaled = scaler.transform(test)
	return scaler, train_scaled, test_scaled
 
# inverse scaling for a forecasted value
def invert_scale(scaler, X, value):
	new_row = [x for x in X] + [value]
	array = numpy.array(new_row)
	array = array.reshape(1, len(array))
	inverted = scaler.inverse_transform(array)
	return inverted[0, -1]

def Ordinal_encoder(train,test):
    encoder = OrdinalEncoder()
    encoder = encoder.fit(train)
    train = train.reshape(train.shape[0], train.shape[1])
    train_encoded = encoder.transform(train)
    test = test.reshape(test.shape[0], test.shape[1])
    test_encoded = encoder.transform(test)
    return encoder, train_encoded, test_encoded


def OrdinalE(train):
    encoder = OrdinalEncoder(dtype=int)
    encoder = encoder.fit(train)
    train = train.reshape(train.shape[0], train.shape[1])
    train_encoded = encoder.transform(train)
    return train_encoded
```

## LSTM model

#### Multivariate model 

```python
import random
import numpy as np
import torch

# multivariate data preparation
from numpy import array
from numpy import hstack
 
# split a multivariate sequence into samples
def split_sequences(sequences, n_steps):
    X, y = list(), list()
    for i in range(len(sequences)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the dataset
        if end_ix > len(sequences):
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequences[i:end_ix, :-1], sequences[end_ix-1, -1]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)
 
# define input sequence
# in_seq1 = array([x for x in range(0,100,10)])
# in_seq2 = array([x for x in range(5,105,10)])
# out_seq = array([in_seq1[i]+in_seq2[i] for i in range(len(in_seq1))])
# # convert to [rows, columns] structure
# in_seq1 = in_seq1.reshape((len(in_seq1), 1))
# in_seq2 = in_seq2.reshape((len(in_seq2), 1))
# out_seq = out_seq.reshape((len(out_seq), 1))
# horizontally stack columns
dataset = hstack((train_x,train_y))
```

```python
class MV_LSTM(torch.nn.Module):
    def __init__(self,n_features,seq_length):
        super(MV_LSTM, self).__init__()
        self.n_features = n_features
        self.seq_len = seq_length
        self.n_hidden = 64 # number of hidden states
        self.n_layers = 2 # number of LSTM layers (stacked)
    
        self.l_lstm = torch.nn.LSTM(input_size = n_features, 
                                 hidden_size = self.n_hidden,
                                 num_layers = self.n_layers, 
                                 batch_first = True,
                                 dropout=0.5)
        # according to pytorch docs LSTM output is 
        # (batch_size,seq_len, num_directions * hidden_size)
        # when considering batch_first = True
        self.l_linear = torch.nn.Linear(self.n_hidden, 1)
        
    
    def init_hidden(self, batch_size):
        # even with batch_first = True this remains same as docs
        hidden_state = torch.zeros(self.n_layers,batch_size,self.n_hidden)
        cell_state = torch.zeros(self.n_layers,batch_size,self.n_hidden)
        
        return (hidden_state, cell_state)
    
    
    def forward(self, x, hidden): 
                        
        batch_size, seq_len, _ = x.size()
        
        lstm_out, hidden = self.l_lstm(x,hidden)
        # lstm_out(with batch_first = True) is 
        # (batch_size,seq_len,num_directions * hidden_size)
        # for following linear layer we want to keep batch_size dimension and merge rest       
        # .contiguous() -> solves tensor compatibility error
        
#         x = torch.tensor(x,dtype=torch.float32)    
        x = lstm_out.contiguous().view(batch_size,seq_len,-1)

        
        return self.l_linear(x)   
```

```python
n_features = 17 # this is number of parallel inputs
n_timesteps = 28 # this is number of timesteps
n_timesteps1 = 16 
n_timesteps2 = 12
batch_size = 32
num_unique = 3049
num_train = 2439
num_test = 610

# X: 3 input dimensions are: (samples, time steps, and features).
# y: 2D Array of dimensions: D1=[samples,D2=timesteps].

X_train, y_train = train_x.reshape(n_timesteps,num_train, train_x.shape[1]).transpose(1,0,2),\
        train_y.values.reshape(n_timesteps,num_train).transpose(1,0)
print(X_train.shape, y_train.shape)


X_test, y_test = test_x.reshape(n_timesteps,num_test,test_x.shape[1]).transpose(1,0,2),\
        test_y.values.reshape(n_timesteps,num_test).transpose(1,0)
print(X_test.shape, y_test.shape)

X_train_nonorm = train_x_nonorm.reshape(n_timesteps,num_train, train_x_nonorm.shape[1]).transpose(1,0,2)
print(X_train_nonorm.shape)

X_test_nonorm = test_x_nonorm.reshape(n_timesteps,num_test,test_x_nonorm.shape[1]).transpose(1,0,2)
print(X_test_nonorm.shape)

# create NN
mv_net = MV_LSTM(n_features,n_timesteps)

criterion = torch.nn.MSELoss() # reduction='sum' created huge loss value
optimizer = torch.optim.Adam(mv_net.parameters(), lr=0.01)
```

Tensor Format: The predictors (X) are 3D arrays of dimensions D1=[samples,D2=timesteps,D3=features], where D1 is the length of values, D1 is the number of time steps (lags), and D3 is the number of predictors (1 if univariate, or n if multivariate) <br>
Outcomes/Targets: (y) is a 2D Array of dimensions: D1=[samples,D2=timesteps]. <br>
Training/Testing: The training and testing length must be evenly divisible (e.g., training lengthtesting length must be an integer). <br>
Batch Size: Represents the number of training examples in one forward/backward pass of the RNN prior to a weight update. The batch size must be evenly divisible into both the training and the testing lengths. <br>
Time Steps: The time step is the number of lags included in the training/testing set. <br>
Epochs: The epochs represent the total number of forward/backward pass iterations. Higher number of epochs tends to improve model performance, unless overfitting occurs, however, itâ€™s more computationally intense. <br>

```python
def train_model(
  model,
  batch_size,
  train_data,
  train_labels,
  test_data=None,
  test_labels=None
):
    loss_fn = torch.nn.MSELoss(reduction='sum')
    optimiser = torch.optim.Adam(model.parameters(), lr=0.01)
    num_epochs = 20
    train_hist = np.zeros(num_epochs)
    test_hist = np.zeros(num_epochs)
    preds_list = []
    best_loss = float('inf')
    
    for t in range(num_epochs):       
        train_loss_list = []  
        test_loss_list = []  
        for b in range(0,len(train_data),batch_size):
            optimiser.zero_grad()
            
            inpt = train_data[b:b+batch_size,:,:]
            target = train_labels[b:b+batch_size]    
            x_batch = torch.tensor(inpt,dtype=torch.float32)    
            y_batch = torch.tensor(target,dtype=torch.float32)
    
            hidden = model.init_hidden(x_batch.size(0))
            
            y_pred = model(x_batch, hidden).squeeze(-1)
            preds_list.append(torch.flatten(y_pred))
            
            loss = criterion(y_pred.float(), y_batch)
            
            loss.backward()
            optimiser.step()
            train_loss_list.append(loss.item())
            
#             print(x_batch[0], y_pred.float()[0], y_batch[0]) 
#             _ = input()
            
        
#             print(f'Batch {b} train loss: {loss.item()}')
           
        mean_train_loss = mean(train_loss_list)
        
        train_hist[t] = mean_train_loss
        print(f'Epoch {t} train loss: {mean_train_loss}') 
        
#         if mean_train_loss < best_loss:
#             mean_train_loss = best_loss
#             torch.save(model, 'best_model.pt')
# #             torch.save(model.state_dict(), 'best_model.pt')   
                
        if test_data is not None:
            for bt in range(0,len(test_data),batch_size):
                inpt_test = test_data[bt:bt+batch_size,:,:]
                target_test = test_labels[bt:bt+batch_size]    
        
                x_batch_test = torch.tensor(inpt_test,dtype=torch.float32)    
                y_batch_test = torch.tensor(target_test,dtype=torch.float32)
                hidden = model.init_hidden(x_batch_test.size(0))
                with torch.no_grad():
                    y_test_pred = model(x_batch_test, hidden).squeeze(-1)
                    test_loss = criterion(y_test_pred.float(), y_batch_test)
                    
                test_loss_list.append(test_loss.item())
            
            mean_test_loss = mean(test_loss_list)
            test_hist[t] = mean_test_loss
            print(f'Epoch {t} test loss: {mean_test_loss}')
              
            if mean_test_loss < best_loss:
                mean_test_loss = best_loss
                torch.save(model, 'best_model.pt')
#             torch.save(model.state_dict(), 'best_model.pt')   
              
    return model.eval(), train_hist, test_hist
```
